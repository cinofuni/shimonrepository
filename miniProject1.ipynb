 A NumPy array representing a sinusoidal signal with a frequency of 5Hz,
sampled every 0.01 seconds.


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#Here i did a a NumPy array representing a sinusoidal signal with a frequency of 5Hz,
#sampled every 0.01 seconds
x = np.arange(0,1,0.01)
signal = np.sin(2 * np.pi * x * 5)

#printing the results
plt.title('Sigmoid Signal')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.grid(True)
plt.plot(signal)

```

A new numpy array representing the same signal. The length of the new array
should be approximately 20% of the original one.


```python
#Here i did a a new numpy array representing the same signal. The length of the new array
#should be approximately 20% of the original one.
reducedSignal = signal[::5]
xReducedSignal = x[::5]
plt.scatter(x[::5], reducedSignal)
xSmooth = np.linspace(x[0], x[-1], 1000)
from numpy import interp
smoothSignal = np.interp(xSmooth, xReducedSignal, reducedSignal)
plt.title('Reduced Sigmoid Signal')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.grid(True)
plt.plot(xSmooth, smoothSignal)




```

• Plot the price of all the laptops

```python
#uploading the file
df = pd.read_csv('laptop_price - dataset.csv')

#finding the price column and saving it
priceColumn = df['Price (Euro)']

#creating the plot
plt.figure(figsize=(10, 6))
plt.plot(df["Price (Euro)"], marker='o')
plt.bar(df.index, priceColumn)
plt.title('Price of All Laptops')
plt.xlabel('Laptops')
plt.ylabel('Price (Euro)')
plt.grid(True)
plt.show()
```

• Which company has on average the most expensive laptop? What is the average
laptop price for each company?


```python
import pandas as pd

#grouping the prices and company together in one calculated average
companyPrices = df.groupby('Company')['Price (Euro)'].mean()

#finding thr company with the most expensive price on average
mostExpensiveCompany = companyPrices.idxmax()
mostExpensivePrice = companyPrices.max()

#printing the results
print(companyPrices)
print(f"\nThe company with the most expensive laptops on average is {mostExpensiveCompany} with an average price of {mostExpensivePrice}")
```

• Find the different types of Operating systems present in the data - under the column
name "OpSys".
o Please note - there are operating systems that are the same systems and just
written differently in the column - please fix them to be uniform.

```python
#mapping together all the OpSystems that got named in different names
osMapping = {
    'Windows 7': 'Windows',
    'Windows 10' : 'Windows',
    'Windows 10 S' : 'Windows',
    'Mac OS X' : 'Mac',
    'macOS' : 'Mac',
}
#appling the mapping to the OpSys column
df['OpSys'] = df['OpSys'].replace(osMapping)

#counting the amount of every OpSystem
osCount = df['OpSys'].value_counts()

#printing the results
print("\nCount of each Operating System after standardization:")
print(osCount)
print("\nUnique Operating Systems after standardization:")
print(osCount.index.tolist())

```

• Plot for each of the operating system types the distribution of the prices, so that the
number of plots equals to the number of unique operating systems.

```python
#checking for unique OpSystems
unique_os = df['OpSys'].unique()
print(f"Unique Operating Systems: {unique_os}")

#creating a plot for each OpSystem and printing it
for os in unique_os:

    osPrices = df[df['OpSys'] == os]['Price (Euro)']

    plt.figure(figsize=(8, 5))
    plt.hist(osPrices, bins=20,)
    plt.title(f"Price Distribution for {os}")
    plt.xlabel('Price (Euro)')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

What is the relationship between RAM and computer price? add an adequate plot to
support your findings.

```python
#extracting the RAM and price columns
ramColumn = df.iloc[:, 8]
priceColumn = df['Price (Euro)']
df['RAM'] = ramColumn

#creating a regression line
slope, intercept = np.polyfit(df['RAM'], priceColumn, 1)
regressionLine = slope * df['RAM'] + intercept

#calculating the correlation
correlation = df['RAM'].corr(priceColumn)

#printing the results
plt.figure(figsize=(10, 6))
plt.scatter(df['RAM'], priceColumn,)
plt.plot(df['RAM'], regressionLine, linestyle='-',)
plt.title('Relationship Between RAM and Price')
plt.xlabel('RAM (GB)')
plt.ylabel('Price (Euro)')
plt.grid(True)
plt.show()
print(f'The correlation Between RAM and Price is: {correlation:.2f}')
```
